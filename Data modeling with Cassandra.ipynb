```python
import cassandra 
from cassandra.cluster import Cluster 
import csv 
```

```python
try :
    cluster = Cluster(['127.0.0.1'])
    session = cluster.connect()
except Exception as e :
    print(e)
```

```python
#Create KeySpace Sparkfiy
try :
    create_query = " create keyspace if not exists SPARKFIY with \
    replication = \
    { 'class' : 'SimpleStrategy' ,'replication_factor' : '1'} "
    session.execute(create_query)
    print('created')
except Exception as e :
    print(e)
```

```python
session.execute('use SPARKFIY')
```

```python
#Query 1 
#Give me the artist, song title and song's length in the music app history 
#was heard during sessionId = 338, and itemInSession = 4
#Attr = artist , song_title , float(song_len) , sessionId (PK) , iteminSession (PK)
#session.execute('truncate table Songs_Q1')
```

```python
#Table for Q1 Songs_Q1 

try :
    session.execute('''create table if not exists songs_q1 (
                     artist text ,
                     song_title text,
                     song_len float,
                     sessionId text,
                     itemInSession text ,
                     primary key(sessionId , itemInSession))''')
    print('table created')
except Exception as e :
    print(e)
```

```python
#Insertion of Query1 

with open('event_datafile_new.csv','r') as f :
    reader = csv.reader(f) 
    next(reader)
    for row in reader :
        artist , fname , gender , iis , lname , length , level , loc , sid , song , uid = row
        try :
            query = '''insert into songs_q1(artist,song_title,song_len,sessionId,itemInSession)
                        values(%s,%s,%s,%s,%s)'''
            session.execute(query,(artist,song,float(length),sid,iis))
            print('Row inserted')
        except Exception as e :
            print(e)
```

```python
#Execution of Q1 
#Give me the artist, song title and song's length in the music app history 
#was heard during sessionId = 338, and itemInSession = 4

try :
    result = session.execute('''select song_title,artist,song_len from songs_q1 
                              where sessionId = '338' and itemInSession = '4' ''')
    for row in result :
        print("The song is ( {} )\nArtist ( {} )\nLength ( {} )".format(row.song_title,row.artist,row.song_len),'\n')
except Exception as e :
    print(e)
```

```python
#Query2 
#Give me only the following: name of artist, song (sorted by itemInSession) and user (first and last name)
#for userid = 10, sessionid = 182
#attr (artist , song , iteminsession , fname , lname , uid , sid )
```

```python
#Table for Q2 Songs_Q2

try :
    session.execute('''create table if not exists songs_q2 (
                     artist text ,
                     song_title text,
                     sessionId int,
                     itemInSession int ,
                     fname text ,
                     lname text ,
                     userid int ,
                     primary key((userid,sessionid),itemInSession) )''')
    print('table created')
except Exception as e :
    print(e)
```

```python
#Insertion of Query1 

with open('event_datafile_new.csv','r') as f :
    reader = csv.reader(f) 
    next(reader)
    for row in reader :
        artist , fname , gender , iis , lname , length , level , loc , sid , song , uid = row
        try :
            query = '''insert into songs_q2(artist,song_title,sessionId,itemInSession,fname,lname,userid)
                        values(%s,%s,%s,%s,%s,%s,%s)'''
            session.execute(query,(artist,song,int(sid),int(iis),fname,lname,int(uid)))
            print('Row inserted')
        except Exception as e :
            print(e)
```

```python
#Execution of Q2
#Give me only the following: name of artist, song (sorted by itemInSession) and user (first and last name)
#for userid = 10, sessionid = 182

try :
    result = session.execute('''select artist,song_title,itemInSession,fname,lname from songs_q2 
                              where userid = 10 and sessionid = 182 order by itemInSession desc''')
    for row in result :
        print("{} :\nArtist ( {} )\nThe song is ( {} )\nName ( {} {} )".format(row.iteminsession,row.artist,row.song_title,row.fname,row.lname),'\n')
except Exception as e :
    print(e)
```

```python
#Query3 
#Give me every user name (first and last) in my music app history 
#who listened to the song 'All Hands Against His Own'
#attr (fname , lname , song pk , uid pk)
```

```python
#Table for Q3 Songs_Q3

try :
    session.execute('''create table if not exists songs_q3 (
                     song  text ,
                     fname text ,
                     lname text ,
                     userid int ,
                     primary key(song , userid) )''')
    print('table created')
except Exception as e :
    print(e)
```

```python
#Insertion of Query3

with open('event_datafile_new.csv','r') as f :
    reader = csv.reader(f) 
    next(reader)
    for row in reader :
        artist , fname , gender , iis , lname , length , level , loc , sid , song , uid = row
        try :
            query = '''insert into songs_q3(song,fname,lname,userid)
                        values(%s,%s,%s,%s)'''
            session.execute(query,(song,fname,lname,int(uid)))
            print('Row inserted')
        except Exception as e :
            print(e)
```

```python
#Execution of Q3 
#Give me every user name (first and last) in my music app history 
#who listened to the song 'All Hands Against His Own'

try :
    result = session.execute('''select fname , lname from songs_q3
                              where song = 'All Hands Against His Own' ''')
    for row in result :
        print("User name = {} {}".format(row.fname , row.lname))
except Exception as e :
    print(e)
```
